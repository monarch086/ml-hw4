```python
import pandas as pd
```

```python
train_df = pd.read_csv('../EDA/train_ml.csv')
train_df
```

```python
test_df = pd.read_csv('../EDA/test_ml.csv')

test_df = test_df.drop(columns='Unnamed: 0')

test_df
```

```python

train_data = train_df.iloc[:, 7:]
train_labels = train_df.iloc[:, 1:7]
```

```python
from sklearn.model_selection import cross_validate
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.multioutput import MultiOutputClassifier
from sklearn.neural_network import MLPClassifier

# model = MLPClassifier()
model = MultiOutputClassifier(LogisticRegression(C=0.05));

# FITTING WITH FULL PCA SCALED DATA
test = cross_validate(model, train_data, train_labels, scoring='roc_auc')
print(test['test_score'])
print(f'{round(test["test_score"].mean() * 100, 2)}%')
model.fit(train_data, train_labels)
```

```python
preds = model.predict(test_df.iloc[:, 1:4])
```

```python
final_df = test_df.drop(columns=test_df.iloc[:, 1:(4)])

final_df.iloc[:, 1:] = preds

final_df
```

```python
final_df.to_csv('../data/submission_multioutput_kneighbors.csv', index=False)
```
